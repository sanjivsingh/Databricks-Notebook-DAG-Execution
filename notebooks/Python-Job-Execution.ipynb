```
import logging
import os
import inspect
import time
import random

class Job():

    task_status_map = {}

    TRIGGERED = "TRIGGERED"
    SKIPPED="SKIPPED"
    COMPLETED="COMPLETED"

    def __update_task_end_status(self,task_name,task_status=COMPLETED):
        self.task_status_map[task_name] = task_status

    def __update_task_start_status(self,task_name,task_status=TRIGGERED):
        self.task_status_map[task_name] = task_status
            
    def executeflow(self):
        self.step1()
        print(self.task_status_map)

    def step1(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED):
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next
            self.step2()
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step2(self.SKIPPED)

    def step2(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED): 
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next
            self.step3()
            self.step4()
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step3(self.SKIPPED)
            self.step4(self.SKIPPED)

    def step3(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED):
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            status = random.randint(10,20)%2
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next 
            if status == 0: 
                self.step5() 
                self.step7(self.SKIPPED) 
            else:
                self.step7()
                self.step5(self.SKIPPED) 
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step5(self.SKIPPED) 
            self.step7(self.SKIPPED) 


    def step4(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED): 
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next
            self.step9() 
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step9(self.SKIPPED) 

    def step5(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED):
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next
            self.step6() 
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step6(self.SKIPPED) 

    def step6(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED): 
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next
            self.step10() 
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step10(self.SKIPPED) 

    def step7(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED): 
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next
            self.step8() 
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step8(self.SKIPPED) 

    def step8(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED):
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next
            self.step10() 
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step10(self.SKIPPED) 

    def step9(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED): 
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next
            self.step10() 
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step10(self.SKIPPED) 

    def step10(self, trigger_type=TRIGGERED):
        previous_steps = ['step6','step8','step9']
        for previous_step in previous_steps:
             if previous_step not in self.task_status_map:
                  return
                 
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED):
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
            # trigger next
            self.step11() 
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------
            self.step11(self.SKIPPED)

    def step11(self, trigger_type=TRIGGERED):
        self.__update_task_start_status(inspect.currentframe().f_code.co_name,trigger_type)
        if(trigger_type != self.SKIPPED):
            #-------------------------- 
            print(f"executed {inspect.currentframe().f_code.co_name}")
            time.sleep(random.randint(0,2))
            #-------------------------- 
            self.__update_task_end_status(inspect.currentframe().f_code.co_name)
        else:
            #-------------------------- 
            print(f"skiped {inspect.currentframe().f_code.co_name}")
            #--------------------------

```
